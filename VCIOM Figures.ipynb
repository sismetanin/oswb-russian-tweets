```python
import pandas as pd

df = pd.read_excel('schastie.xls', index_col=0, skiprows=1)
df
```

```python
list(df.iloc[0])
```

```python
from matplotlib import pyplot as plt
from collections import Counter
import matplotlib.ticker as mtick
from matplotlib.dates import MonthLocator, DateFormatter

%matplotlib inline

plt.style.use('ggplot')

fig = plt.gcf()
fig.set_size_inches(9, 4)

x_dates = df.columns[1:-1]
x = [i for i in range(len(x_dates))]



plt.plot(x, list(df.iloc[5])[1:-1], label='Happinnes Index')
plt.plot(x, list(df.iloc[0]+df.iloc[1])[1:-1], label='Positive Affect')
plt.plot(x, list(df.iloc[2]+df.iloc[3])[1:-1], label='Negative Affect')

plt.plot(x, list(df.iloc[0])[1:-1], '--', alpha=0.5, label='Yes')
plt.plot(x, list(df.iloc[1])[1:-1], '--', alpha=0.5, label='Rather Yes')
plt.plot(x, list(df.iloc[2])[1:-1], '--', alpha=0.5, label='Rather No')
plt.plot(x, list(df.iloc[3])[1:-1], '--', alpha=0.5, label='No')
plt.plot(x, list(df.iloc[4])[1:-1], '--', alpha=0.5, label="Don't know")


ax = fig.axes[0]
ax.set_xticklabels([date.strftime("%b %Y") for date in x_dates], rotation=90)
ax.set_xticks(x)
plt.legend(ncol=4,loc='upper center',  bbox_to_anchor=(0.5, 1.23))#,loc='center', bbox_to_anchor=(0.5, -0.2))
plt.xlim(min(x), max(x))
plt.ylim(0, 100)
plt.show()
fig.savefig('vciom_happiness.png', dpi=180, bbox_inches='tight')
```

```python
df_latex = pd.DataFrame({
    'Date': [date.strftime("%b %Y") for date in x_dates],
    'Yes': list(df.iloc[0])[1:-1],
    'Rather Yes': list(df.iloc[1])[1:-1],
    'Rather No': list(df.iloc[2])[1:-1],
    'No': list(df.iloc[3])[1:-1],
    "Don't know": list(df.iloc[4])[1:-1],
    'Happinnes Index': list(df.iloc[5])[1:-1]
})
```
